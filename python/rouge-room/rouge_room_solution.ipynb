## Part A

### <span style="font-family:monospace;">Solution</span>

````python
def draw_room(w, h, x, y):
    # print the top wall
    print("+" + w * "-" + "+")
    for i in range(h):
        if i == y:
            print("|" + x*"." + "@" + (w - x - 1)*"." + "|")
        else:
            print("|" + w*"." + "|")
    # print the bottom wall
    print("+" + w * "-" + "+")
````

````python
rooms = [(2, 2, 0, 1)]
for i, t in enumerate(rooms):
    print(f"Room {i + 1}")
    draw_room(*t)
````

### <u><span style="font-family:monospace;">Solution</span></u>

````python
class Room:
    def __init__(self, width, height, floor_marker='.', player_marker='@', wall_marker='|', corner_marker='+'):
        self.width = width
        self.height = height
        self.x = 0
        self.y = 0
        self.floor_marker= floor_marker
        self.player_marker = player_marker
        self.wall_marker = wall_marker
        self.corner_marker = corner_marker
        

    def set_player_marker(self, marker):
        if isinstance(marker, str):
            self.set_player_marker = marker

    def set_wall_marker(self, marker):
        if isinstance(marker, str):
            self.wall_marker = marker

    def set_corner_marker(self, marker):
        if isinstance(marker, str):
            self.corner_marker = marker

    def draw(self):
        # print the top wall
        print(self.corner_marker + self.width * "-" + self.corner_marker)
        for i in range(self.height):
            if i != self.y:
                print(self.wall_marker + self.width * self.floor_marker + self.wall_marker)
            else:
                print(self.wall_marker + self.x * self.floor_marker \
                        + self.player_marker + (self.width - self.x - 1)*self.floor_marker \
                        + self.wall_marker)
        print(self.corner_marker + self.width * "-" + self.corner_marker)

    def left(self):
        if (self.x > 0):
            self.x -= 1
        else:
            print("Cannot move left further!")
    
    def right(self):
        if (self.x < self.width - 1):
            self.x += 1
        else:
            print("Cannot move right further!")

    def up(self):
        if (self.y > 0):
            self.y -= 1
        else:
            print("Cannot move up further!")

    def down(self):
        if (self.y < self.height - 1):
            self.y += 1
        else:
            print("Cannot move down further!")
````

````python
from IPython.display import clear_output
import time
````

````python
room = Room(8, 6)
counter = 0
while True:
    if counter == 0:
        room.draw()
    
    s = input()
    if s == 'a':
        room.left()
    if s == 'd':
        room.right()
    if s == 'w':
        room.up()
    if s == 's':
        room.down()
    if s == 'x':
        print("all done")
        break
    time.sleep(0.5)    
    if counter != 0:
        room.draw()
    clear_output(wait=True)
    counter += 1
````
